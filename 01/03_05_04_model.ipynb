```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
iris = datasets.load_iris()

(iris_train_ftrs, iris_test_ftrs, iris_train_tgt,
iris_test_tgt) = train_test_split(iris.data, iris.target,
test_size=.25)
print("Oblik polja značajki za treniranje:", iris_train_ftrs.shape)
print("Oblik polja značajki za testiranje:", iris_test_ftrs.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=3)
fit = knn.fit(iris_train_ftrs, iris_train_tgt)
preds = fit.predict(iris_test_ftrs)
print("3NN accuracy:",
metrics.accuracy_score(iris_test_tgt, preds))
```
