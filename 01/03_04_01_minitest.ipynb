```python
from sklearn import metrics
import numpy as np

tocni_odgovori = np.array([True, True, False, True])
studentovi_odgovori = np.array([True, True, True, True])
ispravno = tocni_odgovori == studentovi_odgovori
print(ispravno)
brojcano_ispravno = ispravno.sum() # True == 1
print("Ručni izračun:", brojcano_ispravno / len(tocni_odgovori))
print("sklearn izračun:", metrics.accuracy_score(tocni_odgovori, studentovi_odgovori))
```
