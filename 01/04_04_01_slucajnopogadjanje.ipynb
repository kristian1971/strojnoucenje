```python
import numpy as np
tgt = np.array([3, 5, 8, 10, 12, 15])
num_guesses = 10
results = []
for g in range(num_guesses):
  guess = np.random.uniform(low=tgt.min(), high=tgt.max())
  print(guess)
  total_dist = np.sum((tgt - guess)**2)
  print(tgt)
  print(total_dist)
  results.append((total_dist, guess))
best_guess = sorted(results)[0][1]
print("_____________")
print(best_guess)
```
