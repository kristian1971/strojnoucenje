```python
import numpy as np
podaci = np.array([1, 3, 5, 8, 11, 13, 15])
print("bez anomalija")
print(np.mean(podaci), np.median(podaci))
podaci_sa_anomalijom = np.array([1, 3, 5, 8, 11, 13, 40])
print("sa anomalijom")
print("%5.2f" % np.mean(podaci_sa_anomalijom), np.median(podaci_sa_anomalijom))
```

```python
udaljenost = np.array([2.0, 4.0, 4.0])
blizina = 1.0 / udaljenost # element-by-element division
tezina = blizina / np.sum(blizina) # normalize sum to one
cijena = np.array([180000, 200000, 220000])
mean = np.mean(cijena)
wgt_mean = np.dot(cijena, tezina)
print("Mean:", mean)
print("Weighted Mean:", wgt_mean)
```
